```python
# Importing libraries
import numpy as np
import pandas as pd
import ast
import plotly.express as px
from plotly import graph_objects as go
```

```python
# Loading csv file
df = pd.read_csv('/content/flipkart_com-ecommerce_sample.csv')
```

```python
df.head(5)
```

```python
#checking for null values
df.isnull().sum()
```

```python
df.tail(5)
```

```python
# removing last row
df = df.drop(df.index[-1])
```

```python
df.isnull().sum()
```

```python
# Replacing retail_price and discounted_price with medians
df['retail_price'].fillna(df['retail_price'].median(), inplace=True)
df['discounted_price'].fillna(df['discounted_price'].median(), inplace=True)
```

```python
# Fetching the discount percentage column
x=df['retail_price']-df['discounted_price']
y=(x/df['retail_price'])*100
df['discount_price']=y
```

```python
# Coverting timestamp into separate date and time cloumns
df['timestamp']=pd.to_datetime(df['crawl_timestamp'])
df['date']=df['timestamp'].dt.date
df['time']=df['timestamp'].dt.time
df.drop(['crawl_timestamp','timestamp'],axis=1,inplace=True)
df.head(2)
```

```python
# Obtaing the main category
def get_main_category(category):
    # Remove brackets and quotes
    if isinstance(category, str):
        category = category.strip("[]'\"")
        if ' >> ' in category:
            return category.split(' >> ')[0]
    return category

# Apply function to the 'product_category_tree' column to create 'main_category'
df['main_category'] = df['product_category_tree'].apply(get_main_category)
```

```python
df.drop('product_category_tree',axis=1,inplace=True)
```

```python
df
```

```python
# Renaming column name
df.rename(columns={'discount_price': 'discount_percent'}, inplace=True)
```

```python
# Finding top brands and top products
n=10
top_products=pd.DataFrame(df['main_category'].value_counts()[:n]).reset_index()
top_products.rename(columns={'index':'Top_products','main_category':'Total_count'}, inplace=True)

n=10
top_brands=pd.DataFrame(df['brand'].value_counts()[:n]).reset_index()
top_brands.rename(columns={'index':'Top_brands','brand':'Total_count'}, inplace=True)

```

```python
top_products.columns
```

```python
top_products.rename(columns={'Total_count': 'Top_products'}, inplace=True)
top_products.rename(columns={'count': 'Total_count'}, inplace=True)
top_brands.rename(columns={'Total_count': 'Top_brands'}, inplace=True)
top_brands.rename(columns={'count': 'Total_count'}, inplace=True)
```

```python
top_brands
```

```python
from plotly.subplots import make_subplots

label1=top_products['Top_products']
value1=top_products['Total_count']
label2=top_brands['Top_brands']
value2=top_brands['Total_count']

fig_both = make_subplots(
    rows=1,
    cols=2,
    specs=[[{'type':'domain'}, {'type':'domain'}]]
)

# Add pie charts to subplots
fig_both.add_trace(go.Pie(labels=label1, values=value1, name='Top Products'), row=1, col=1)
fig_both.add_trace(go.Pie(labels=label2, values=value2, name='Top Brands'), row=1, col=2)

# Update traces to be donut charts
fig_both.update_traces(hole=0.5, hoverinfo="label+percent+name")

# Update layout
fig_both.update_layout(
    title_text="Top Products and Brands Distribution",
    annotations=[
        dict(text='Product', x=0.18, y=0.5, font_size=20, showarrow=False),
        dict(text='Brand', x=0.82, y=0.5, font_size=20, showarrow=False)
    ]
)

# Show the figure
fig_both.show()
```

```python
# Finding out which product is giving highest discount
df_discount=df[df['discount_percent']>80]
df_discount=df_discount.dropna()
df_discount["brand"].replace("FlashBlush","Flash Blush",inplace=True)
max_discount=pd.DataFrame(df_discount.groupby('brand')[['discount_percent']].mean().sort_values(by=['discount_percent'],ascending=False).reset_index())
```

```python
max_discount
```

```python
px.bar(max_discount, x='brand', y='discount_percent', color='brand',color_discrete_sequence=px.colors.qualitative.Plotly)
```

```python
df_customer=df.groupby('uniq_id')[['discounted_price']].sum().sort_values(by=['discounted_price'],ascending=[False])

# Top 20 customers spending the most
list1=df_customer[:20]
list1.reset_index(inplace=True)

fig = px.bar(list1, x='uniq_id', y='discounted_price', color='discounted_price', color_continuous_scale=px.colors.diverging.Tealrose)

# Show the figure
fig.show()
```

```python
# 5 star rating

total_prod=len(df['pid'])
total_ratings=len(df[df['product_rating']!='No rating available'])
top_ratings=len(df[df['product_rating']=='5'])
df_funnel_1= dict(
    number=[total_prod,total_ratings,top_ratings],
    stage=['Total Products','Products with ratings','Products with 5 star rating'])

funnel_1_fig=px.funnel(df_funnel_1,x='number',y='stage')
funnel_1_fig.show()
```

```python
# 5 star products/brands
rating_5=pd.DataFrame(df.loc[df['product_rating']=='5'])
top_product_type=rating_5['main_category'].value_counts()
top_brand_type=rating_5['brand'].value_counts()

# top 5 products
df_top_product=pd.DataFrame(top_product_type[:5].reset_index())
df_top_product.rename(columns={'index':'top_prod'},inplace=True)
df_top_product.drop('main_category',axis=1,inplace=True)

# top 5 brands
df_top_brand=pd.DataFrame(top_brand_type[:5].reset_index())
df_top_brand.rename(columns={'index':'top_brands'},inplace=True)
df_top_brand.drop('brand',axis=1,inplace=True)

df_product_brand_rate5=pd.concat([df_top_product,df_top_brand],axis=1)
```

```python


# Assuming df is your original DataFrame
# Drop rows where 'product_rating' is 'No rating available'
df.drop(df.index[df['product_rating'] == 'No rating available'], inplace=True)

# Create a DataFrame with the count of each rating
ratings = pd.DataFrame(df['product_rating'].value_counts()).sort_values(by='product_rating',ascending=True).reset_index()


# Plotting the result
data = ratings
x = data['product_rating']
y = data['count']

figdot2 = go.Figure()
figdot2.add_trace(go.Scatter(
    x=x,
    y=y,
    marker=dict(color='crimson', size=12),
    mode='markers',
    name='ratings'
))

figdot2.update_layout(title='Ratings vs Counts',
                      xaxis_title='Ratings',
                      yaxis_title='Counts')

figdot2.update_xaxes(showline=True, linewidth=1, linecolor='black', mirror=True)
figdot2.update_yaxes(showline=True, linewidth=1, linecolor='black', mirror=True)
figdot2.show()
```
